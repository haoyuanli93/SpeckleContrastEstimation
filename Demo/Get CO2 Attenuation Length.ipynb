# This notebook serves as a demo of how to get the CO2 attenuation length

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")

from ContrastEstimation import util, IntensityEstimation
```

```python
# Get the molecule structure
"""
One can either define the molecule structure or 
use a PDB file to load existing structure
or just select some molecules
that I have saved to the module
"""
my_molecule = util.get_molecule_from_pdb("../pdb_files/co2.pdb")

# Show the source code of how I have defined a molecule for calculation.
util.show_formated_molecule_structure(my_molecule)
```

```python
# Incident photon energy
photon_energy_keV_list = np.array([6000.00, 6645.40, 7360.22, 8151.93, 9028.80, 10000.0, ])/ 1000.

# Get the attenuation length for various density
density_list = np.array([0.1, 0.3, 0.5, 0.7, 0.9])  # unit: g/cm^3

# Get the corresponding attenuation legnth
attenuation_length_list = []
for energy in photon_energy_keV_list:
    attenuation_length_list.append([])
    for density in density_list:
        attenuation_length_list[-1].append(IntensityEstimation.get_attenuation_length_cm(
            molecule_structure_list=[my_molecule,],
            partial_density_list=[density,],
            photon_energy_keV = energy) )

attenuation_length_list = np.array(attenuation_length_list)
```

```python
fig, axis = plt.subplots(ncols=1, nrows=1)


for x in range(5):
    axis.plot(density_list, attenuation_length_list[x] * 1e4,
              label="{:.2f} keV".format(photon_energy_keV_list[x]))

axis.set_xlabel("Density (g/cm^3)")
axis.set_ylabel("Atten Length (um)")
axis.legend()
plt.show()
```

# Compare with the data obtained from the website:

that of the website.

```python
attenuation_length_from_website = np.array(
    [
        [4353.35, 5964.24, 8179.69, 11226.0, 15412.6, 21149.0, ],
        [1451.12, 1988.08, 2726.56, 3741.99, 5137.54, 7049.66, ],
        [870.669, 1192.85, 1635.94, 2245.19, 3082.52, 4229.80, ],
        [621.907, 852.035, 1168.53, 1603.71, 2201.80, 3021.28, ],
        [483.705, 662.694, 908.855, 1247.33, 1712.51, 2349.89, ],
    ]
)
attenuation_length_from_website = attenuation_length_from_website.T
```

```python
fig, axis = plt.subplots(ncols=1, nrows=1)


for x in range(5):
    axis.plot(density_list, 
              (attenuation_length_list[x] * 1e4 - attenuation_length_from_website[x]) / attenuation_length_from_website[x],
              label="{:.2f} keV".format(photon_energy_keV_list[x]))

axis.set_xlabel("Density (g/cm^3)")
axis.set_ylabel("Relative Error")
axis.legend()
plt.show()
```
